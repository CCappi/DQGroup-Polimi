```python
import pandas as pd
```

```python
newexp_path = "../NewExp/"
fullimp_path = "../Full_ImpExp_ML/"
final_files_path = "../Experiments/final_files/"

new_exp_categorical = ['name', 'column_name', 'n_tuples', 'missing_perc', 'constancy',
       'imbalance', 'uniqueness', 'unalikeability', 'entropy', 'density',
       'mean_char', 'std_char', 'skewness_char', 'kurtosis_char', 'min_char',
       'max_char', 'ml_method', 'impute_standard', 'impute_mode',
       'impute_random', 'impute_knn', 'impute_mice',
       'impute_logistic_regression', 'impute_random_forest', 'impute_kproto',
       'impute_xgb_imputer', 'impute_catboost', 'impute_rfi',
       'impute_autoimpute', 'impute_mlp']

new_exp_numerical = ['name', 'column_name', 'n_tuples', 'missing_perc', 'uniqueness', 'min',
       'max', 'mean', 'median', 'std', 'skewness', 'kurtosis', 'mad', 'iqr',
       'p_min', 'p_max', 'k_min', 'k_max', 's_min', 's_max', 'entropy',
       'density', 'ml_algorithm', 'impute_standard', 'impute_mean',
       'impute_median', 'impute_random', 'impute_knn', 'impute_mice',
       'impute_linear_regression', 'impute_random_forest', 'impute_cmeans',
       'impute_expectation_maximization', 'impute_soft_imputer',
       'impute_xgb_imputer', 'impute_catboost', 'impute_rfi',
       'impute_autoimpute', 'impute_gain', 'impute_vae', 'impute_mlp']

partial_newexp_categorical = ['name', 'column_name', 'n_tuples', 'missing_perc', 'constancy',
       'imbalance', 'uniqueness', 'unalikeability', 'entropy', 'density',
       'mean_char', 'std_char', 'skewness_char', 'kurtosis_char', 'min_char',
       'max_char', 'ml_method','impute_xgb_imputer', 'impute_catboost', 'impute_rfi', 'impute_mlp']

partial_newexp_numerical = ['name', 'column_name', 'n_tuples', 'missing_perc', 'uniqueness', 'min',
       'max', 'mean', 'median', 'std', 'skewness', 'kurtosis', 'mad', 'iqr',
       'p_min', 'p_max', 'k_min', 'k_max', 's_min', 's_max', 'entropy',
       'density', 'ml_algorithm', 'impute_expectation_maximization', 'impute_soft_imputer',
       'impute_xgb_imputer', 'impute_catboost', 'impute_rfi',
       'impute_autoimpute', 'impute_gain', 'impute_vae', 'impute_mlp']

# Unify dataframes df_newexp with the missing processed dataset of df_fullimp
for i in range(1, 9):
       print("Experiment ", i)
       df_newexp_categorical = pd.read_csv(newexp_path + "experiment_" + str(i) + "_categorical.csv", header=None, names=partial_newexp_categorical, skiprows=1, usecols=range(len(partial_newexp_categorical)))
       df_newexp_numerical = pd.read_csv(newexp_path + "experiment_" + str(i) + "_numerical.csv", header=None, names=partial_newexp_numerical, skiprows=1, usecols=range(len(partial_newexp_numerical)))
       # Retrieve dataframe from csv with a given header
       df_fullimp_categorical = pd.read_csv(fullimp_path + "experiment_" + str(i) + "_categorical.csv", header=None, names=partial_newexp_categorical, skiprows=1)
       df_fullimp_numerical = pd.read_csv(fullimp_path + "experiment_" + str(i) + "_numerical.csv", header=None, names=partial_newexp_numerical, skiprows=1)
       print("NewExp categorical shape: ", df_newexp_categorical.shape)
       print("NewExp numerical shape: ", df_newexp_numerical.shape)
       print("FullImp categorical shape: ", df_fullimp_categorical.shape)
       print("FullImp numerical shape: ", df_fullimp_numerical.shape)

       # Drop experiments from df_fullimp that are already done in NewExp
       for row in df_fullimp_categorical.itertuples():
              if row.name in df_newexp_categorical['name'].values and row.column_name in df_newexp_categorical['column_name'].values:
                     df_fullimp_categorical.drop(row[0], inplace=True)
       for row in df_fullimp_numerical.itertuples():
              if row.name in df_newexp_numerical['name'].values and row.column_name in df_newexp_numerical['column_name'].values:
                     df_fullimp_numerical.drop(row[0], inplace=True)
    
       #Unify dataframes df_newexp_categorical and df_fullimp_categorical by concatenating rows 
       df_unified_categorical = pd.concat([df_newexp_categorical, df_fullimp_categorical])
       df_unified_categorical.sort_values(by=['name'], inplace=True)
       #Unify dataframes df_newexp_numerical and df_fullimp_numerical by concatenating rows 
       df_unified_numerical = pd.concat([df_newexp_numerical, df_fullimp_numerical])
       df_unified_numerical.sort_values(by=['name'], inplace=True)

       # Save the unified dataframes to csv
       df_unified_categorical.to_csv("../Experiments/ml_methods_exp/NewImp/experiment_" + str(i) + "_categorical.csv", index=False)
       df_unified_numerical.to_csv("../Experiments/ml_methods_exp/NewImp/experiment_" + str(i) + "_numerical.csv", index=False)


# Unify experiments from final_files with old imputation techniques together with experiments from ml_methods_exp/NewImp with new imputation techniques
# Drop experiments from final_files that are not present in experiments from ml_methods_exp/NewImp
for i in range(1, 9):
       print("Experiment ", i)
       df_final_categorical = pd.read_csv(final_files_path + "experiment_" + str(i) + "_categorical.csv")
       df_final_numerical = pd.read_csv(final_files_path + "experiment_" + str(i) + "_numerical.csv")
       df_newimp_categorical = pd.read_csv("../Experiments/ml_methods_exp/NewImp/experiment_" + str(i) + "_categorical.csv")
       df_newimp_numerical = pd.read_csv("../Experiments/ml_methods_exp/NewImp/experiment_" + str(i) + "_numerical.csv")
       print("Final categorical shape: ", df_final_categorical.shape)
       print("Final numerical shape: ", df_final_numerical.shape)
       print("NewImp categorical shape: ", df_newimp_categorical.shape)
       print("NewImp numerical shape: ", df_newimp_numerical.shape)

       for row in df_final_categorical.itertuples():
              if row.name not in df_newimp_categorical['name'].values or row.column_name not in df_newimp_categorical['column_name'].values:
                     df_final_categorical.drop(row[0], inplace=True)
       for row in df_final_numerical.itertuples():
              if row.name not in df_newimp_numerical['name'].values or row.column_name not in df_newimp_numerical['column_name'].values:
                     df_final_numerical.drop(row[0], inplace=True)
       
       df_final_categorical.sort_values(by=['name'], inplace=True)
       df_final_numerical.sort_values(by=['name'], inplace=True)

       # Add the columns of df_newimp_categorical to df_final_categorical that are not present in df_final_categorical
       for column in df_newimp_categorical.columns:
              if column not in df_final_categorical.columns:
                     df_final_categorical[column] = df_newimp_categorical[column]
       for column in df_newimp_numerical.columns:
              if column not in df_final_numerical.columns:
                     df_final_numerical[column] = df_newimp_numerical[column]

       # Save the unified dataframes to csv
       df_final_categorical.to_csv("../Experiments/ml_methods_exp/All_Imp/experiment_" + str(i) + "_categorical.csv", index=False)
       df_final_numerical.to_csv("../Experiments/ml_methods_exp/All_Imp/experiment_" + str(i) + "_numerical.csv", index=False)
```

```python
# Merge Deep Learning methods experiments
deep_learning_path = "../DeepLearningExps/"
deep_learning_server_path = "../DeepLearningExps_Server/"

new_exp_categorical = ['name', 'column_name', 'n_tuples', 'missing_perc', 'constancy',
       'imbalance', 'uniqueness', 'unalikeability', 'entropy', 'density',
       'mean_char', 'std_char', 'skewness_char', 'kurtosis_char', 'min_char',
       'max_char', 'ml_method', 'impute_standard', 'impute_mode',
       'impute_random', 'impute_knn', 'impute_mice',
       'impute_logistic_regression', 'impute_random_forest', 'impute_kproto',
       'impute_xgb_imputer', 'impute_catboost', 'impute_rfi',
       'impute_autoimpute', 'impute_mlp']

new_exp_numerical = ['name', 'column_name', 'n_tuples', 'missing_perc', 'uniqueness', 'min',
       'max', 'mean', 'median', 'std', 'skewness', 'kurtosis', 'mad', 'iqr',
       'p_min', 'p_max', 'k_min', 'k_max', 's_min', 's_max', 'entropy',
       'density', 'ml_algorithm', 'impute_standard', 'impute_mean',
       'impute_median', 'impute_random', 'impute_knn', 'impute_mice',
       'impute_linear_regression', 'impute_random_forest', 'impute_cmeans',
       'impute_expectation_maximization', 'impute_soft_imputer',
       'impute_xgb_imputer', 'impute_catboost', 'impute_rfi',
       'impute_autoimpute', 'impute_gain', 'impute_vae', 'impute_mlp']


for i in range(1, 9):
      print("Experiment ", i)
      df_deep_learning_categorical = pd.read_csv(deep_learning_path + "experiment_" + str(i) + "_categorical.csv", header=None, names=new_exp_categorical, skiprows=1)
      df_deep_learning_server_categorical = pd.read_csv(deep_learning_server_path + "experiment_" + str(i) + "_categorical.csv", header=None, names=new_exp_categorical, skiprows=1)
      df_deep_learning_numerical = pd.read_csv(deep_learning_path + "experiment_" + str(i) + "_numerical.csv", header=None, names=new_exp_numerical, skiprows=1)
      df_deep_learning_server_numerical = pd.read_csv(deep_learning_server_path + "experiment_" + str(i) + "_numerical.csv", header=None, names=new_exp_numerical, skiprows=1)
  
      # Drop experiments from df_deep_learning that are already done in df_deep_learning_server
      for row in df_deep_learning_server_categorical.itertuples():
            if row.name in df_deep_learning_categorical['name'].values and row.column_name in df_deep_learning_categorical['column_name'].values:
                  df_deep_learning_server_categorical.drop(row[0], inplace=True)
      
      for row in df_deep_learning_server_numerical.itertuples():
            if row.name in df_deep_learning_numerical['name'].values and row.column_name in df_deep_learning_numerical['column_name'].values:
                  df_deep_learning_server_numerical.drop(row[0], inplace=True)
      
      #Unify dataframes df_deep_learning and df_deep_learning_server by concatenating rows 
      df_unified_deep_learning_categorical = pd.concat([df_deep_learning_categorical, df_deep_learning_server_categorical])
      df_unified_deep_learning_categorical.sort_values(by=['name'], inplace=True)
      # Drop column impute_autoimpute from df_unified_deep_learning_categorical
      df_unified_deep_learning_categorical.drop(columns=['impute_autoimpute'], inplace=True)
      
      df_unified_deep_learning_numerical = pd.concat([df_deep_learning_numerical, df_deep_learning_server_numerical])
      df_unified_deep_learning_numerical.sort_values(by=['name'], inplace=True)

      # Save the unified dataframe to csv
      df_unified_deep_learning_categorical.to_csv("../Experiments/dl_methods_exp/experiment_" + str(i) + "_categorical.csv", index=False)
      df_unified_deep_learning_numerical.to_csv("../Experiments/dl_methods_exp/experiment_" + str(i) + "_numerical.csv", index=False)
```

```python
# Unify Experiments from dl_methods_exp and ml_methods_exp/All_Imp
dl_methods_path = "../Experiments/dl_methods_exp/"
ml_methods_path = "../Experiments/ml_methods_exp/All_Imp/"

for i in range(1, 9):
    print("Experiment ", i)
    df_dl_categorical = pd.read_csv(dl_methods_path + "experiment_" + str(i) + "_categorical.csv")
    df_dl_numerical = pd.read_csv(dl_methods_path + "experiment_" + str(i) + "_numerical.csv")
    df_ml_categorical = pd.read_csv(ml_methods_path + "experiment_" + str(i) + "_categorical.csv")
    df_ml_numerical = pd.read_csv(ml_methods_path + "experiment_" + str(i) + "_numerical.csv")

    final_df_categorical = pd.concat([df_ml_categorical, df_dl_categorical]).sort_values(by=['name', 'column_name'])
    final_df_numerical = pd.concat([df_ml_numerical, df_dl_numerical]).sort_values(by=['name', 'column_name'])

    final_df_categorical.to_csv("../Experiments/all_methods_exp/experiment_" + str(i) + "_categorical.csv", index=False)
    final_df_numerical.to_csv("../Experiments/all_methods_exp/experiment_" + str(i) + "_numerical.csv", index=False)
```
